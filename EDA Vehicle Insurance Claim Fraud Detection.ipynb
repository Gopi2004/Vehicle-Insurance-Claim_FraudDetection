```python
# Summary statistics for numerical variables
summary_stats = data.describe()
print(summary_stats)
```

```python
# Histogram for Age
plt.figure(figsize=(8, 6))
plt.hist(data['Age'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()

# Box plot for Age by FraudFound_P
plt.figure(figsize=(8, 6))
sns.boxplot(x='FraudFound_P', y='Age', data=data)
plt.title('Boxplot of Age by FraudFound_P')
plt.show()

# Bar plot for MaritalStatus
plt.figure(figsize=(8, 6))
sns.countplot(x='MaritalStatus', data=data)
plt.title('Count of MaritalStatus')
plt.show()
```

```python
# Cross-tabulation between MaritalStatus and FraudFound_P
cross_tab = pd.crosstab(data['MaritalStatus'], data['FraudFound_P'])
print(cross_tab)

# Stacked bar plot for MaritalStatus and FraudFound_P
stacked_bar = cross_tab.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('MaritalStatus vs. FraudFound_P')
plt.xlabel('MaritalStatus')
plt.ylabel('Count')
plt.legend(title='FraudFound_P')
plt.show()
```

```python
# Box plot for Age to detect outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['Age'])
plt.title('Boxplot of Age (Outlier Detection)')
plt.show()
```

```python
# Counting the number of occurrences for each marital status
marital_counts = data['MaritalStatus'].value_counts()

# Plotting the pie chart with custom colors
plt.figure(figsize=(8, 6))
colors = ['gold', 'lightskyblue', 'lightcoral', 'lightgreen']
plt.pie(marital_counts, labels=marital_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Marital Status Distribution')
plt.show()
```

```python
# Counting the number of occurrences for each fault attribution
fault_counts = data['Fault'].value_counts()

# Plotting the pie chart with custom colors
plt.figure(figsize=(8, 6))
colors = ['lightcoral', 'lightgreen']
plt.pie(fault_counts, labels=fault_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Fault Distribution')
plt.show()
```

```python
# Counting the number of occurrences for police report filed
police_counts = data['PoliceReportFiled'].value_counts()

# Plotting the pie chart with updated custom colors
plt.figure(figsize=(8, 6))
colors = ['#66c2a5', '#fc8d62']  # Custom colors for better visibility
plt.pie(police_counts, labels=police_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Police Report Filed Distribution')
plt.show()
```
